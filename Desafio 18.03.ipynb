```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dias_semana = ["Dom", "Seg", "Ter", "Qua", "Qui", "Sex", "Sáb"]

temperaturas = np.random.randint(20, 36, size=7)

dados_clima = pd.DataFrame({
    'Dia': dias_semana,
    'Temp': temperaturas
})

media = dados_clima['Temp'].mean()
t_max = dados_clima['Temp'].max()
t_min = dados_clima['Temp'].min()

print(f"Relatório Semanal:")
print(f"-> Média térmica: {media:.1f}°C")
print(f"-> Extremos: Máxima de {t_max}°C / Mínima de {t_min}°C")

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].bar(dados_clima['Dia'], dados_clima['Temp'], color='skyblue', edgecolor='navy')
ax[0].set_title("Variação Diária (Barras)")
ax[0].set_ylabel("Graus Celsius (°C)")

ax[1].plot(dados_clima['Dia'], dados_clima['Temp'], marker='o', linestyle='--', color='tomato')
ax[1].set_title("Tendência da Semana (Linha)")
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
